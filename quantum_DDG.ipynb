### **2. Task Definition and Motivation**

**Objective:**

We aim to optimize the parameters \( \vec{\theta} \) of a variational quantum circuit \( U(\vec{\theta}) \) to minimize a cost function \( C(\vec{\theta}) \) related to a specific task, such as classification, state preparation, or energy minimization in quantum chemistry.

**Mathematical Formulation:**

\[
\min_{\vec{\theta}} C(\vec{\theta}) = \sum_{i} \langle \psi_i | U^\dagger(\vec{\theta})\, O\, U(\vec{\theta}) | \psi_i \rangle
\]

- **\( |\psi_i\rangle \)**: Input quantum states.
- **\( U(\vec{\theta}) \)**: Parameterized quantum circuit.
- **\( O \)**: Observable operator representing the measurement of interest.

**Motivation:**

Optimizing VQCs is essential for leveraging quantum computers to solve complex problems more efficiently than classical computers. By fine-tuning the parameters \( \vec{\theta} \), we can harness quantum phenomena like superposition and entanglement to achieve computational advantages in tasks such as:

- **Quantum Machine Learning**: Classifying quantum data or enhancing classical machine learning models with quantum components.
- **Quantum Chemistry**: Finding ground state energies of molecules.
- **Optimization Problems**: Solving complex optimization tasks faster than classical algorithms.

### **4. Algorithm Details**

#### **Inputs and Outputs**

- **Inputs:**
  - **Quantum States (\( |\psi_i\rangle \))**: The dataset of quantum states.
  - **Initial Parameters (\( \vec{\theta}^{(0)} \))**: Starting values for the circuit parameters.
  - **Observable (\( O \))**: The measurement operator defining the cost function.
  - **Hyperparameters:**
    - Learning rate \( \eta \).
    - Adam optimizer parameters \( \beta_1, \beta_2, \epsilon \).
    - Maximum iterations \( T \) or convergence tolerance.

- **Outputs:**
  - **Optimized Parameters (\( \vec{\theta}^* \))**: Parameters that minimize the cost function.
  - **Minimum Cost Value (\( C(\vec{\theta}^*) \))**: The lowest achieved cost.

#### **Algorithm Steps**

**Initialization:**

1. **Set Initial Parameters:**
   - Choose \( \vec{\theta}^{(0)} \) randomly or based on prior knowledge.

2. **Initialize Optimizer Variables:**
   - First moment estimates \( \vec{m}^{(0)} = \vec{0} \).
   - Second moment estimates \( \vec{v}^{(0)} = \vec{0} \).
   - Iteration counter \( t = 0 \).

**Optimization Loop:**

Repeat until convergence or maximum iterations \( T \) are reached:

1. **Increment Iteration Counter:**
   - \( t = t + 1 \).

2. **For Each Parameter \( \theta_k \):**

   a. **Compute Gradient Using Parameter-Shift Rule:**

      \[
      \frac{\partial C}{\partial \theta_k} = \frac{1}{2} \left[ C\left( \vec{\theta}^{(t-1)} + \frac{\pi}{2} \vec{e}_k \right) - C\left( \vec{\theta}^{(t-1)} - \frac{\pi}{2} \vec{e}_k \right) \right]
      \]

      - \( \vec{e}_k \) is the unit vector with 1 at position \( k \) and 0 elsewhere.

   b. **Update First Moment Estimate (Momentum):**

      \[
      m_k^{(t)} = \beta_1 m_k^{(t-1)} + (1 - \beta_1) \left( \frac{\partial C}{\partial \theta_k} \right)
      \]

   c. **Update Second Moment Estimate (Velocity):**

      \[
      v_k^{(t)} = \beta_2 v_k^{(t-1)} + (1 - \beta_2) \left( \frac{\partial C}{\partial \theta_k} \right)^2
      \]

   d. **Compute Bias-Corrected Estimates:**

      \[
      \hat{m}_k^{(t)} = \frac{m_k^{(t)}}{1 - \beta_1^t}, \quad \hat{v}_k^{(t)} = \frac{v_k^{(t)}}{1 - \beta_2^t}
      \]

   e. **Update Parameter:**

      \[
      \theta_k^{(t)} = \theta_k^{(t-1)} - \eta \frac{\hat{m}_k^{(t)}}{\sqrt{\hat{v}_k^{(t)}} + \epsilon}
      \]

3. **Check for Convergence:**

   - If \( |C(\vec{\theta}^{(t)}) - C(\vec{\theta}^{(t-1)})| < \text{tolerance} \), stop the optimization.
   - Else, proceed to the next iteration.

**Final Output:**

- Optimized parameters \( \vec{\theta}^* = \vec{\theta}^{(t)} \).
- Minimum cost value \( C(\vec{\theta}^*) \).

#### **Memory Requirements**

- **Classical Memory:**
  - Store parameters \( \vec{\theta} \), first moments \( \vec{m} \), and second moments \( \vec{v} \).
  - Memory usage scales linearly with the number of parameters \( N \): \( \mathcal{O}(N) \).

- **Quantum Memory:**
  - Quantum processors operate on quantum states without storing them due to the no-cloning theorem.
  - Memory is used per circuit execution but doesn't accumulate over iterations.
  - Quantum memory usage is determined by the number of qubits and circuit depth, not by the number of iterations.

### **5. Mathematical Proof of Correctness**

Let's prove that the algorithm correctly computes the gradients and converges toward minimizing the cost function.

#### **Step 1: Correctness of the Parameter-Shift Rule**

**Goal:** Demonstrate that the parameter-shift rule provides the exact gradient of the cost function with respect to each parameter \( \theta_k \).

**Assumptions:**

- The parameterized gate \( U_k(\theta_k) \) is of the form \( e^{-i \theta_k G_k} \), where \( G_k \) is a Hermitian operator with eigenvalues \( \pm r \).

**Proof:**

1. **Express the Cost Function:**

   \[
   C(\vec{\theta}) = \sum_i \langle \psi_i | U^\dagger(\vec{\theta})\, O\, U(\vec{\theta}) | \psi_i \rangle
   \]

2. **Compute the Derivative w.r.t \( \theta_k \):**

   \[
   \frac{\partial C}{\partial \theta_k} = \sum_i \frac{\partial}{\partial \theta_k} \langle \psi_i | U^\dagger(\vec{\theta})\, O\, U(\vec{\theta}) | \psi_i \rangle
   \]

3. **Apply the Chain Rule:**

   \[
   \frac{\partial C}{\partial \theta_k} = \sum_i \left( \langle \psi_i | \frac{\partial U^\dagger(\vec{\theta})}{\partial \theta_k} O U(\vec{\theta}) | \psi_i \rangle + \langle \psi_i | U^\dagger(\vec{\theta}) O \frac{\partial U(\vec{\theta})}{\partial \theta_k} | \psi_i \rangle \right)
   \]

4. **Compute the Derivatives of \( U(\vec{\theta}) \):**

   - The derivative of the unitary operator is:

     \[
     \frac{\partial U(\vec{\theta})}{\partial \theta_k} = -i U_1 \cdots U_{k-1} G_k U_k(\theta_k) U_{k+1} \cdots U_N
     \]

   - Similarly,

     \[
     \frac{\partial U^\dagger(\vec{\theta})}{\partial \theta_k} = i U_N^\dagger \cdots U_{k+1}^\dagger U_k^\dagger(\theta_k) G_k U_{k-1}^\dagger \cdots U_1^\dagger
     \]

5. **Simplify the Expression:**

   - Let \( |\phi_i(\vec{\theta})\rangle = U(\vec{\theta}) |\psi_i\rangle \).
   - Then,

     \[
     \frac{\partial C}{\partial \theta_k} = -i \sum_i \left( \langle \phi_i(\vec{\theta}) | G_k^\dagger O | \phi_i(\vec{\theta}) \rangle - \langle \phi_i(\vec{\theta}) | O G_k | \phi_i(\vec{\theta}) \rangle \right)
     \]

6. **Use Hermiticity of \( G_k \):**

   - Since \( G_k^\dagger = G_k \), we have:

     \[
     \frac{\partial C}{\partial \theta_k} = -i \sum_i \langle \phi_i(\vec{\theta}) | [G_k, O] | \phi_i(\vec{\theta}) \rangle
     \]

7. **Parameter-Shift Rule Application:**

   - For \( G_k \) with eigenvalues \( \pm r \), the gradient can be expressed as:

     \[
     \frac{\partial C}{\partial \theta_k} = r \left[ C\left( \vec{\theta} + \frac{\pi}{2r} \vec{e}_k \right) - C\left( \vec{\theta} - \frac{\pi}{2r} \vec{e}_k \right) \right]
     \]

   - For \( r = \frac{1}{2} \), the shift is \( \pi \).
   - For \( r = 1 \) (common case), the shift is \( \frac{\pi}{2} \).

**Conclusion:**

- The parameter-shift rule provides an exact method to compute the gradient \( \frac{\partial C}{\partial \theta_k} \) by evaluating the cost function at shifted parameter values.
- This ensures accurate gradient computation without numerical approximation errors.

### **6. Memory Complexity Analysis**

Let's analyze the memory requirements of the algorithm in detail.

#### **Classical Memory Complexity**

- **Parameters Storage (\( \vec{\theta} \)):**
  - Stores \( N \) real numbers.
  - Complexity: \( \mathcal{O}(N) \).

- **First and Second Moments (\( \vec{m}, \vec{v} \)):**
  - Each stores \( N \) real numbers.
  - Combined complexity: \( \mathcal{O}(N) \).

- **Gradients (\( \vec{g} \)):**
  - Stores \( N \) real numbers per iteration.
  - Since we can overwrite the gradient vector each iteration, the memory doesn't accumulate.
  - Complexity: \( \mathcal{O}(N) \).

- **Total Classical Memory:**
  - The total classical memory required scales linearly with the number of parameters:
    \[
    \text{Memory}_\text{classical} = \mathcal{O}(N)
    \]

#### **Quantum Memory Complexity**

- **Quantum Circuit Execution:**
  - Quantum memory is used per execution of the circuit.
  - The number of qubits required is determined by the problem size (number of qubits \( Q \)) and circuit depth.

- **Memory Accumulation:**
  - Quantum states are not stored across iterations due to the no-cloning theorem.
  - Each circuit execution is independent in terms of memory.

- **Total Quantum Memory:**
  - The quantum memory requirement is:
    \[
    \text{Memory}_\text{quantum} = \mathcal{O}(Q)
    \]
  - Independent of the number of parameters \( N \) or iterations \( T \).

#### **Comparison with Classical Algorithms**

- **Classical Gradient Descent:**
  - May require storing large datasets and gradients.
  - For large-scale problems, classical methods can have memory complexities higher than \( \mathcal{O}(N) \).

- **Quantum Advantage:**
  - The quantum algorithm maintains linear memory scaling with respect to parameters.
  - Avoids storing large intermediate data, which can be a bottleneck in classical algorithms.

### Problem Definition
We aim to optimize a variational quantum circuit, a common model for quantum neural networks, where the circuit parameters are tuned to minimize a cost function. This cost function is typically related to the output of the quantum circuit when applied to a set of input quantum states.

### Input States
- **Quantum States (\(|\psi_i\rangle\))**: These are the input states to the quantum neural network. Each state \(|\psi_i\rangle\) is prepared as an input to the quantum circuit.
- **Parameterized Quantum Circuit (\(U(\vec{\theta})\))**: The circuit is defined by a set of parameters \(\vec{\theta}\), which are adjusted during training. \(U(\vec{\theta})\) operates on \(|\psi_i\rangle\) to produce an output state.

### Objective
The goal is to adjust \(\vec{\theta}\) such that the expectation value of a certain observable \(O\) on the output state \(U(\vec{\theta})|\psi_i\rangle\) is optimized. Mathematically, we often minimize a cost function defined as:
\[ C(\vec{\theta}) = \sum_i \langle \psi_i | U^\dagger(\vec{\theta}) O U(\vec{\theta}) | \psi_i \rangle \]
This cost function can be related to classification accuracy, fidelity against a target state, or other metrics depending on the specific quantum machine learning task.

### Gradient Estimation Using the Parameter-Shift Rule
To optimize the circuit, we need to calculate gradients of the cost function with respect to each parameter \(\theta_k\). Using the parameter-shift rule, the gradient can be estimated as:
\[ \frac{\partial C}{\partial \theta_k} \approx \frac{1}{2} \left[ C(\theta_k + \frac{\pi}{2}) - C(\theta_k - \frac{\pi}{2}) \right] \]
Where:
- \(C(\theta_k \pm \frac{\pi}{2})\) is the cost function evaluated with the parameter \(\theta_k\) shifted by \(\pm \frac{\pi}{2}\), keeping other parameters fixed.

### Parallel Execution Across Quantum Processors
Each quantum processor can independently compute the gradient with respect to one or more parameters by performing the following steps:
1. **Prepare Input State**: Each processor prepares the same set of input states \(|\psi_i\rangle\).
2. **Apply Quantum Circuit**: Perform the quantum operations \(U(\theta_k \pm \frac{\pi}{2})\) on the states.
3. **Measure Observable**: Measure the observable \(O\) on the output states and compute the cost function values \(C(\theta_k + \frac{\pi}{2})\) and \(C(\theta_k - \frac{\pi}{2})\).
4. **Compute Gradient**: Calculate the gradient for \(\theta_k\) using the results of the measurements.

### Algorithm Integration with Adam Optimization and Efficient Memory Usage
With gradients estimated, the Adam optimization algorithm can be employed to update the parameters \(\vec{\theta}\). This involves updating the running averages of the gradients (first moment) and their squared values (second moment) and using these moments to adjust \(\vec{\theta}\). To manage memory efficiently, we will directly apply the gradient updates without storing intermediate gradients.

### Enhanced Pseudocode with Detailed Steps

#### Inputs:
- **Quantum States \( |\psi_i\rangle \)**: These are the input quantum states over which the variational quantum circuit is evaluated.
- **Quantum Circuit \( U(\vec{\theta}) \)**: A parameterized quantum circuit defined by parameters \( \vec{\theta} \).
- **Observable \( O \)**: The observable whose expectation value we seek to optimize.

#### Outputs:
- **Optimized Parameters \( \vec{\theta} \)**: The parameters of the quantum circuit that minimize the cost function based on the given observable.

#### Algorithm Steps:

```pseudo
Initialize parameters θ with random values
Initialize first moment m and second moment v to zero
Initialize learning rate η

repeat until convergence or maximum iterations are reached:
    for each parameter θ_k in parallel across quantum processors:
        Prepare input states |ψ_i⟩
        
        // Parameter-shift rule to compute gradients
        Set θ_k_plus = θ_k + π/2
        Set θ_k_minus = θ_k - π/2
        
        // Compute the cost function for both shifted parameters
        Cost_plus = Measure the cost function for U(θ_k_plus)
        Cost_minus = Measure the cost function for U(θ_k_minus)
        
        // Compute the gradient using the parameter-shift rule
        Gradient_k = (Cost_plus - Cost_minus) / 2
        
        // Apply Grover's algorithm to amplify the gradient
        Amplified_Gradient_k = Apply Grover's algorithm to Gradient_k
        
        // Update first and second moments for Adam optimization
        m_k = β1 * m_k + (1 - β1) * Amplified_Gradient_k
        v_k = β2 * v_k + (1 - β2) * (Amplified_Gradient_k)^2
        
        // Correct bias in first and second moment estimates
        m_hat_k = m_k / (1 - β1^t)
        v_hat_k = v_k / (1 - β2^t)
        
        // Update parameter
        θ_k = θ_k - η * m_hat_k / (sqrt(v_hat_k) + ε)
    
    // Optionally, adjust learning rate η if necessary

// Output the optimized parameters
Output optimized parameters θ
```

### Explanation of Key Components:

1. **Quantum State Preparation**: Prepare the quantum state \( |\psi_i\rangle \) before each gradient estimation.
2. **Gradient Estimation via Parameter-Shift Rule**: Adjust the parameter \( \theta_k \) by \( \pm \pi/2 \), measure the cost functions, and calculate the gradient.
3. **Grover's Algorithm for Gradient Amplification**: Apply Grover's algorithm to amplify the calculated gradient, improving the optimization process.
4. **Direct Gradient Application**: Directly use the amplified gradients to update the first and second moments for Adam optimization, avoiding exponential memory requirements.
5. **Adam Optimization**: Update parameters using the Adam algorithm, which adjusts parameters based on the first and second moments of past gradients.
6. **Parameter Updates**: Update parameters with the Adam optimization rules, using bias-corrected versions of the accumulated moments.

This refined pseudocode ensures efficient memory usage while leveraging parallel processing and Grover's algorithm for gradient amplification, enhancing the optimization of variational quantum circuits.

````python
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit.circuit import Parameter
from qiskit_aer import Aer
import matplotlib.pyplot as plt
````

### Complexity Analysis of Classical Backpropagation

For comparison, let's analyze the complexity of classical backpropagation in a neural network.

### Problem Definition
In classical neural networks, backpropagation is used to compute the gradient of the loss function with respect to each parameter in the network. The complexity analysis involves the following steps:

1. **Forward Pass**
2. **Backward Pass**
3. **Parameter Update**

### 1. Forward Pass
During the forward pass, the input data is passed through each layer of the network to compute the output.

- **Complexity**: If the network has \(L\) layers and each layer \(l\) has \(n_l\) neurons, the complexity is determined by the matrix multiplications and activation functions.
- **Overall Complexity**: \(\mathcal{O}\left(\sum_{l=1}^{L} n_{l-1} n_l \right)\)

### 2. Backward Pass
The backward pass computes the gradients of the loss function with respect to each parameter using the chain rule.

- **Complexity**: The complexity of the backward pass is similar to the forward pass since it involves the same number of matrix multiplications.
- **Overall Complexity**: \(\mathcal{O}\left(\sum_{l=1}^{L} n_{l-1} n_l \right)\)

### 3. Parameter Update
The parameters are updated using the computed gradients. If an optimizer like Adam is used, additional operations for maintaining first and second moment estimates are involved.

- **Complexity**: \(\mathcal{O}(P)\) where \(P\) is the total number of parameters.
- **Overall Complexity**: \(\mathcal{O}(P)\)

### Total Complexity
Combining all these components, the total complexity of classical backpropagation is:

\[ \mathcal{O}\left(\sum_{l=1}^{L} n_{l-1} n_l \right) \]

### Discussion
The classical backpropagation complexity scales with the size of the network and the number of neurons in each layer. For deep networks with many layers and large neuron counts, this can become quite significant.

### Comparison to Quantum Approach

#### Quantum Gradient Estimation with Grover's Algorithm:
- **State Preparation**: \(\mathcal{O}(N)\)
- **Circuit Execution**: \(\mathcal{O}(Nd)\)
- **Gradient Estimation**: \(\mathcal{O}(P\sqrt{N})\)
- **Synchronization**: \(\mathcal{O}(P\log Q)\)
- **Adam Optimization**: \(\mathcal{O}(P)\)

Total Quantum Complexity:
\[ \mathcal{O}(N) + \mathcal{O}(Nd) + \mathcal{O}(P\sqrt{N}) + \mathcal{O}(P\log Q) + \mathcal{O}(P) \]

#### Classical Backpropagation:
\[ \mathcal{O}\left(\sum_{l=1}^{L} n_{l-1} n_l \right) + \mathcal{O}(P) \]

### Key Observations
- The classical approach's complexity is highly dependent on the network's depth and the number of neurons per layer.
- The quantum approach benefits from the quadratic speedup provided by Grover's algorithm for gradient estimation, making it potentially more efficient for large \(N\) (number of input states).
- The quantum approach's complexity includes additional terms for synchronization and parallel execution, which are not present in the classical approach.

### Summary
The classical backpropagation algorithm's complexity is primarily driven by the number of layers and neurons, while the quantum approach leverages parallelism and quantum speedup for gradient estimation. For large datasets (\(N\)), the quantum approach can offer significant advantages, especially when the number of quantum processors \(Q\) can efficiently distribute the computation. However, the practical implementation of quantum algorithms also depends on the available quantum hardware and the overheads associated with quantum state preparation, circuit execution, and measurement.
